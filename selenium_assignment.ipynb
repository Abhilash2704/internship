```python
page_url = ['https://www.azquotes.com/top_quotes.html',
           'https://www.azquotes.com/top_quotes.html?p=2',
           'https://www.azquotes.com/top_quotes.html?p=3',
           'https://www.azquotes.com/top_quotes.html?p=4',
           'https://www.azquotes.com/top_quotes.html?p=5',
           'https://www.azquotes.com/top_quotes.html?p=6',
           'https://www.azquotes.com/top_quotes.html?p=7',
           'https://www.azquotes.com/top_quotes.html?p=8',
           'https://www.azquotes.com/top_quotes.html?p=9',
           'https://www.azquotes.com/top_quotes.html?p=10']

for i in page_url:
    driver=webdriver.Chrome()
    driver.get(i)
    quote_tags = driver.find_elements(By.XPATH,'//a[@class="title"]')
    author_tags = driver.find_elements(By.XPATH,'//div[@class="author"]')
    type_tags = driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for j in quote_tags:
        tags = j.text
        quote.append(tags)
    
    for j in author_tags:
        tags1 = j.text
        author.append(tags1)
    
    for j in type_tags:
        tags2 = j.text
        type_quo.append(tags2)
    
print(len(quote),len(author),len(type_quo))
```
